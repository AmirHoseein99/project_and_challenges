```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
records = []
with open('/content/drive/MyDrive/Colab Notebooks/DataProjects/datasets/parkinsons.data') as f:
  for line in f.read().split('\n'): 
      records.append(line)
```

```python
datas = []
for rec in records[:] : 
  datas.append(rec.split(','))
```

```python
dataset = pd.DataFrame(datas[1:], columns=datas[:1])
dataset
```

```python
digiCols = dataset.columns.drop('name')
dataset[digiCols] = dataset[digiCols].apply(pd.to_numeric, errors='coerce')
```
